```python
from numbers import Number
import math

def run_equiv_format(t1, d1, d2):
    if isinstance(t1,Number):
        t2 = t1*(d2/d1)**1.06
        return format_seconds(t2)
    else:
        return 9999

def run_equiv(t1, d1, d2):
    if isinstance(t1,Number):
        t2 = t1*(d2/d1)**1.06
        return t2/60
    else:
        return 9999

def format_seconds(t2):
    if isinstance(t2,Number):
        minutes = int(math.floor(t2/60))
        seconds = str(int((t2 % 1)*60))
        if len(seconds)==1:
            seconds = '0' + seconds
        return str(minutes) + ':' + seconds
```
